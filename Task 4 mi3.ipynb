```python
import csv
import re

from collections import defaultdict

columns = defaultdict(list)


with open('short_tweets.csv') as f:
    reader = csv.DictReader(f)
    for row in reader:
        for (k,v) in row.items():
            columns[k].append(v) 
list1=[]

with open('short_tweets.csv') as f:
    s = f.read() + '\n'

stri="(^|[^@\w])@(\w{1,15})"

result = re.findall(stri, repr(s))
print("All names appear in the tweets are : ",result)



```

```python
z=re.findall("(^|[^\d\w])\d(\d\w{1,15})", repr(result))
"(^|[^\d\w])\d(\w{1,15})"

print('\n')
print('Name contains digits are ',z)

```

```python
from collections import Counter

words = re.findall(r'\w+',s)

cap_words = [word.upper() for word in words]

word_counts = Counter(cap_words)

print('Most frequent 50 words are : ',word_counts.most_common(50))
print('\n')
```

```python
str2='love'
x=re.findall(str2, repr(s))
str3='hate'
y=re.findall(str3, repr(s))
if len(x)>len(y):
    print('Love appears more.')
else:
    print('Hate appears more.')
```

```python
str2='good'
x=re.findall(str2, repr(s))
str3='bad'
y=re.findall(str3, repr(s))
if len(x)>len(y):
    print('Good appears more.')
else:
    print('Bad appears more.')
print('\n')    
```

```python
two_before_hate = re.findall(r"(?:[a-zA-Z'-]+[^a-zA-Z'-]+){0,2}hate", s)
print('Hate and two words before it : ', two_before_hate )
```

```python
two_after_love = re.findall(r"love(?:[^a-zA-Z'-]+[a-zA-Z'-]+){0,2}", s)
print('\n')    
print('Love and two words after it : ', two_after_love )
```
